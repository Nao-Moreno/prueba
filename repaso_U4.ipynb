```python
import numpy as np  # Biblioteca para cálculos numéricos
import matplotlib.pyplot as plt  # Biblioteca para graficar
from scipy.stats import norm  # Función para trabajar con estadistica

# **Paso 1: Definir los parámetros de la distribución normal**
mu = 50  # Media (centro de la distribución)
sigma = 2  # Desviación estándar (dispersión de los datos)

# **Paso 2: Generar un rango de valores "x" para graficar la distribución**
# Vamos a crear valores en un rango amplio alrededor de la media (4 desviaciones estándar a ambos lados)
x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 1000)
# Aquí, 'mu - 4 * sigma' es el límite inferior y 'mu + 4 * sigma' el superior.
# np.linspace crea 1000 puntos equidistantes en ese rango.
x
```

```python

# **Paso 3: Calcular la densidad de probabilidad para cada valor en "x"**
# Usamos la función norm.pdf() para obtener la densidad de probabilidad de una normal con media mu y desviación sigma.
y = norm.pdf(x, loc=mu, scale=sigma)
y


#Densidad de Probabilidad
#En una variable aleatoria continua, como el peso de los componentes, no podemos
#calcular la probabilidad exacta de que un valor sea exactamente igual a un número
#(por ejemplo, que un peso sea exactamente 50 g). En lugar de eso, usamos una función
#llamada densidad de probabilidad que nos indica qué tan probable es encontrar valores en un rango específico.

#La curva de densidad muestra cómo están distribuidos los valores.
#El área bajo la curva en un intervalo específico representa la probabilidad de que la variable esté en ese intervalo.

#Por ejemplo:
#Si queremos saber la probabilidad de que el peso esté entre 48 g y 52 g, calculamos el área bajo la curva entre esos valores.
```

```python
# Calcular el valor máximo y mínimo de la densidad de probabilidad (y)
max_y = np.max(y)  # Máximo valor de y (altura máxima de la curva)
min_y = np.min(y)  # Mínimo valor de y (altura mínima de la curva)

# Imprimir los resultados
print(f"El valor máximo de y (densidad de probabilidad) es: {max_y:.4f}")
print(f"El valor mínimo de y (densidad de probabilidad) es: {min_y:.6f}")
```

```python
# **Paso 4: Crear la gráfica**
plt.figure(figsize=(8, 6))  # Establece el tamaño de la figura
```

```python
# Graficar la curva de densidad
plt.plot(x, y, color='blue', label=f'N({mu}, {sigma}²)')
# El color azul representa la distribución normal y se incluye una etiqueta que describe la normal.
```

```python
# **Paso 5: Agregar detalles a la gráfica**
plt.title('Distribución Normal ($\\mu=50$, $\\sigma=2$)', fontsize=14)  # Título de la gráfica
plt.xlabel('Peso (g)', fontsize=12)  # Etiqueta del eje X
plt.ylabel('Densidad de Probabilidad', fontsize=12)  # Etiqueta del eje Y

```

```python
# **Paso 6: Marcar la media con una línea vertical**
plt.axvline(mu, color='red', linestyle='--', label='Media ($\\mu$)')
# Línea punteada roja en la posición de la media.
```

```python
# **Paso 7: Sombrar el intervalo $\mu \pm \sigma$**
plt.fill_between(
    x, y,
    where=(x >= mu - sigma) & (x <= mu + sigma),  # Sombrear entre mu - sigma y mu + sigma
    color='green', alpha=0.2, label='$\\mu \\pm \\sigma$ (68%)'
)
# Esto representa el 68% de los datos en el rango de una desviación estándar alrededor de la media.
```

```python
# **Paso 4: Crear la gráfica**
plt.figure(figsize=(8, 6))  # Establece el tamaño de la figura

# Graficar la curva de densidad
plt.plot(x, y, color='blue', label=f'N({mu}, {sigma}²)')
# El color azul representa la distribución normal y se incluye una etiqueta que describe la normal.

# **Paso 5: Agregar detalles a la gráfica**
plt.title('Distribución Normal ($\\mu=50$, $\\sigma=2$)', fontsize=14)  # Título de la gráfica
plt.xlabel('Peso (g)', fontsize=12)  # Etiqueta del eje X
plt.ylabel('Densidad de Probabilidad', fontsize=12)  # Etiqueta del eje Y

# **Paso 6: Marcar la media con una línea vertical**
plt.axvline(mu, color='red', linestyle='--', label='Media ($\\mu$)')
# Línea punteada roja en la posición de la media.

# **Paso 7: Sombrar el intervalo $\mu \pm \sigma$**
plt.fill_between(
    x, y,
    where=(x >= mu - sigma) & (x <= mu + sigma),  # Sombrear entre mu - sigma y mu + sigma
    color='green', alpha=0.2, label='$\\mu \\pm \\sigma$ (68%)'
)
# Esto representa el 68% de los datos en el rango de una desviación estándar alrededor de la media.

# **Paso 8: Agregar una leyenda y mostrar la cuadrícula**
plt.legend(fontsize=12)  # Mostrar la leyenda
plt.grid(alpha=0.4)  # Agregar cuadrícula con un poco de transparencia

# **Paso 9: Mostrar la gráfica**
plt.show()  # Renderizar la gráfica
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python

# 1. Simulación de datos
np.random.seed(42)  # Para reproducibilidad
n_components = 10000  # Número de componentes
mean_weight = 50  # Media de los pesos (en gramos)
std_dev_weight = 2  # Desviación estándar de los pesos
weights = np.random.normal(mean_weight, std_dev_weight, n_components)
weights
```

```python
# Calcular el valor máximo y mínimo de los pesos
max_weight = np.max(weights)  # Máximo valor en los pesos simulados
min_weight = np.min(weights)  # Mínimo valor en los pesos simulados

# Imprimir los resultados
print(f"El peso máximo es: {max_weight:.2f} g")
print(f"El peso mínimo es: {min_weight:.2f} g")
```

```python
# Simular defectos (probabilidad del 2%)
defect_probability = 0.02
defects = np.random.choice([0, 1], size=n_components, p=[1 - defect_probability, defect_probability])

#p=[1 - defect_probability, defect_probability]:
#Este parámetro es una lista que define las probabilidades de seleccionar cada uno de los valores en la lista [0, 1].
#defect_probability es la probabilidad de que un componente tenga defecto (1), mientras que 1 - defect_probability es la
#probabilidad de que un componente no tenga defecto (0). En el código original, defect_probability = 0.02, lo que significa
#que el 2% de los componentes tienen defecto (valor 1), y el 98% restante no tienen defecto (valor 0). Por lo tanto, las probabilidades son:
#Probabilidad de defecto (1): 0.02 (2%)
#Probabilidad de no defecto (0): 0.98 (98%)


# Crear un DataFrame para análisis
data = pd.DataFrame({'Weight': weights, 'Defect': defects})
```

```python
# Imprimir los primeros 50 valores del DataFrame 'data'
print(data.head(50))
```

```python
# Contar cuántos valores hay con 0 (sin defecto) y cuántos con 1 (con defecto)
count_zero = (data['Defect'] == 0).sum()  # Contar cuántos 0s hay
count_one = (data['Defect'] == 1).sum()   # Contar cuántos 1s hay

# Calcular el porcentaje respecto a los 10,000 elementos
percentage_zero = (count_zero / n_components) * 100
percentage_one = (count_one / n_components) * 100

# Imprimir los resultados
print(f"Cantidad de componentes sin defecto (0): {count_zero}")
print(f"Cantidad de componentes con defecto (1): {count_one}")
print(f"Porcentaje de componentes sin defecto: {percentage_zero:.2f}%")
print(f"Porcentaje de componentes con defecto: {percentage_one:.2f}%")
```

La ligera variación con respecto al 2% y al 98% es normal en cualquier simulación probabilística.

#TLC

TLC: Toma 1,000 muestras de tamaño 30 de los datos simulados, calcula la media de cada muestra y grafica la distribución.
La distribución de las medias muestrales sigue una forma normal más estrecha.

```python
# Teorema del Límite Central (TLC)
sample = [np.mean(np.random.choice(weights, size=30, replace=False)) for _ in range(1000)]
sample_means = [np.mean(sample3) for sample3 in sample]
```


Desglosando el código:
for _ in range(1000):

Este es un bucle que se repetirá 1,000 veces. Se utiliza para generar 1,000 muestras aleatorias y calcular la media de cada una de ellas.
La variable _ es un nombre convencional usado cuando el valor de la variable no se necesita, en este caso porque solo nos interesa el número de repeticiones.
np.random.choice(weights, size=30, replace=False):

np.random.choice() es una función de NumPy que selecciona una muestra aleatoria de un conjunto de datos.
weights: Es el array de datos del cual se va a tomar la muestra (en este caso, los pesos de los componentes).

size=30: Esto indica que se seleccionarán 30 elementos de weights para cada muestra.

replace=False: Esto especifica que el muestreo es sin reemplazo, es decir, cada elemento seleccionado en la muestra no puede volver a ser seleccionado. Por lo tanto, cada muestra consistirá en 30 elementos únicos del conjunto original.

#Teoría de la Estimación

```python

confidence_level = 0.95
sampli = np.random.choice(weights, size=30, replace=False)
sampli_mean = np.mean(sampli)
sampli_std = np.std(sampli, ddof=1)
```

Se establece el nivel de confianza en 95%, lo que significa que estamos interesados en obtener un intervalo dentro del cual esperamos que caiga la media de la población con una probabilidad del 95%.


sample = np.random.choice(weights, size=30, replace=False):
Se selecciona una muestra aleatoria de 30 elementos sin reemplazo (replace=False) de los valores de weights (que simula los pesos de los componentes).
La función np.random.choice() selecciona 30 valores de forma aleatoria de la lista de pesos.


sample_mean = np.mean(sample):
Se calcula la media de la muestra usando np.mean(). Esto nos da el valor promedio de los 30 elementos seleccionados aleatoriamente.


sample_std = np.std(sample, ddof=1):

Se calcula la desviación estándar de la muestra con np.std(), usando ddof=1 para aplicar la corrección de Bessel, que ajusta el cálculo de la desviación estándar cuando se trata de una muestra y no de la población completa. Esto se hace para evitar un sesgo hacia la subestimación de la desviación estándar de la población.
La desviación estándar nos da una medida de cuánto varían los valores de la muestra respecto a la media.

```python
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df=len(sampli) - 1) * (sampli_std / np.sqrt(len(sampli)))
print(f"Media muestral: {sampli_mean:.2f}, Intervalo de confianza al 95%: [{sampli_mean - margin_of_error:.2f}, {sampli_mean + margin_of_error:.2f}]")
```

margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df=len(sample) - 1) * (sample_std / np.sqrt(len(sample))):

Se calcula el margen de error usando la distribución t de Student, que se usa cuando no se conoce la desviación estándar de la población y se debe estimar a partir de la muestra.

stats.t.ppf((1 + confidence_level) / 2, df=len(sample) - 1): Esto calcula el valor crítico de la t de Student para un intervalo de confianza del 95%. La función ppf (percent point function) nos da el valor t correspondiente a un nivel de confianza de 95% (es decir, el área total a la izquierda de este valor es 95%).

(1 + confidence_level) / 2 calcula el percentil del 97.5% (ya que un intervalo de confianza del 95% tiene el 2.5% en cada cola de la distribución).

df=len(sample) - 1 define los grados de libertad de la muestra.

SEM: (sample_std / np.sqrt(len(sample))): Esta parte calcula el error estándar de la media, que es la desviación estándar de la muestra dividida por la raíz cuadrada del tamaño de la muestra (30 en este caso).

El producto de estos dos valores da el margen de error, es decir, cuánto se aleja la media muestral de la verdadera media poblacional con un 95% de confianza.

stats.ttest_1samp(sample, mean_weight):

ttest_1samp es una función de la librería scipy.stats que realiza una prueba t de una muestra para evaluar si la media de la muestra es igual a una media conocida (en este caso, mean_weight = 50 gramos).

El primer parámetro sample es la muestra que estamos analizando. Esta muestra se obtiene de los componentes simulados, como en el resto del código, y contiene una cantidad de valores que representan las características (en este caso, el peso de los componentes).
El segundo parámetro mean_weight es el valor con el cual comparamos la media de la muestra. Aquí estamos probando si la media de los pesos de la muestra es igual a 50 gramos.

La prueba t se utiliza para determinar si existe una diferencia significativa entre la media de una muestra y una media conocida o teórica. En este caso, queremos saber si el peso promedio de los componentes (muestra) es significativamente diferente de 50 gramos.

**Dado que el p-valor (0.6328) es mayor que el umbral común de 0.05, podemos concluir que no hay suficiente evidencia para rechazar la hipótesis nula. Esto significa que, basándonos en esta prueba t, la media de la muestra no es significativamente diferente de 50 gramos.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Parámetros de la prueba t
df = 29  # Grados de libertad (n - 1) para una muestra de tamaño 30
t_stat = 0.6328  # Estadístico t obtenido de la prueba t
p_value = 0.4829  # p-valor calculado de la prueba

# Generar valores para la distribución t de Student
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)

# Crear la figura y el eje
plt.figure(figsize=(8, 6))

# Graficar la distribución t
plt.plot(x, y, label=f'Distribución t de Student (df={df})', color='blue')

# Resaltar el área bajo la curva para el p-valor
# Para un p-valor de 0.4924, resaltamos el área desde t hasta el infinito
x_fill = np.linspace(t_stat, 4, 1000)  # Desde t_stat hasta el valor máximo
y_fill = stats.t.pdf(x_fill, df)
plt.fill_between(x_fill, y_fill, color='red', alpha=0.3, label=f'Área p-valor = {p_value:.4f}')

# Marcar el valor t en la gráfica
plt.axvline(x=t_stat, color='green', linestyle='--', label=f'Valor t = {t_stat:.4f}')

# Añadir etiquetas y título
plt.title('Distribución t de Student y p-valor')
plt.xlabel('t')
plt.ylabel('Densidad de probabilidad')
plt.legend()

# Mostrar la gráfica
plt.show()
```

chi_stat = (len(sample) - 1) * (sample_std ** 2) / (std_dev_weight ** 2): Calcula la estadística chi-cuadrado.

Calcula el p-valor utilizando la distribución chi-cuadrado con los grados de libertad correspondientes. Usamos 1 - chi2.cdf porque estamos interesados en el área a la derecha de chi_stat, que nos da la probabilidad de obtener un valor de chi-cuadrado tan extremo o más extremo que el calculado.

print(f"Prueba chi-cuadrado: χ²={chi_stat:.4f}, p-valor={p_value_chi:.4f}"): Muestra el resultado de la prueba chi-cuadrado con el valor χ² y el p-valor.
